```python
def extract_location(file_name):
    
    locations = (
        "ADAMS FAMILY COURT",
        "CHAMBERS",
        "CHURCH AVENUE",
        "2 STATIONS",
        "3 STATIONS",
        "4 STATIONS",
        "5 STATIONS",
        "ART AND LEATHER H.S",
        "BLEECKER",
        "C-3 CENTER",
        "3 HISTORICAL",
    )
    
    loc = [x for x in locations if x in file_name]
    
    return loc
```

```python
import dateutil
from dateutil.parser import *
import datetime

def extract_date(file_name,debug=False):
    
    ss = file_name.split('/')
    ss.reverse()
    
    #the file name will have the date as MM-DD-YY, but in some cases it's only MM-DD or DD
    s1 = ss[0]
    s1 = s1.rstrip('.xls')
    mo = re.search(r'([0-9]+[-][0-9]+[-][0-9]+)',s1) #fuzzy can't handle "ABC 04-21-06 5", so will help by stripping stuff out
    if mo is not None:
        s1 = mo.group(1)
        
    try:
        dd1 = dateutil.parser.parse(s1,fuzzy=True)
    except (ValueError):
        #Are there cases we can reconstruct the datetime?
        dd1 = datetime.date(1900,1,1)
    
    #the directory name will have the date as MMM YYH
    s2 = ss[1] 
    try:
        dd2 = dateutil.parser.parse('1 '+s2) #dateutil to parse 'MAR 06' as 'Mar 2006' not 'Mar 6th'
    except (ValueError):
        dd2 = datetime.date(1900,1,1)
  
    #building the final date, using file name, and supplementing it with a YY from dir name where missing    
    #from dir - range is April 2003 to Jan 2010
    dd = datetime.date(dd2.year if dd1.year == 2013 or dd1.year < 2003 else dd1.year,dd1.month,dd1.day)
    
    #if the dateutil.parser.parse() failed for s1, then construct date from s2, where day=1
    if dd1 == datetime.date(1900,1,1):
        dd = datetime.date(dd2.year,dd2.month,1)
    
    #validation between the month from dir and file
    valid = [False if dd1.month != dd2.month else True]
    
    #hack to clean up dates for path with no dates, including TEMPLATE, CHURCH, HISTORIC
    if dd.year == 1900:
        dd = datetime.date(1900,1,1)
        valid = True
    
    if debug is True:
        return dd, s1, dd1, s2, dd2, valid
    else:
        return dd
```

```python
def cell(val,type):

#    xlrd.XL_CELL_EMPTY 0
#    xlrd.XL_CELL_TEXT 1
#    xlrd.XL_CELL_NUMBER 2
#    xlrd.XL_CELL_DATE 3
#    xlrd.XL_CELL_BOOLEAN 4
#    xlrd.XL_CELL_ERROR 5
#    xlrd.XL_CELL_BLANK 6
    
    if type == xlrd.XL_CELL_EMPTY: v = 0
    elif type == xlrd.XL_CELL_TEXT: v = 0
    elif type == xlrd.XL_CELL_NUMBER: v = val
    elif type == xlrd.XL_CELL_DATE: v = 0
    elif type == xlrd.XL_CELL_BOOLEAN: v = 0
    elif type == xlrd.XL_CELL_ERROR: v = 0
    elif type == xlrd.XL_CELL_BLANK: v = 0
    
    return v
```

```python
def col_index_from_name(sheet, col_name):
    
    candidate_index = -1
    
    # iterate over each column
    
    for iter in range (0,sheet.ncols):
        seq = sheet.col_values(iter)
        # check if all words in the col_name appear in the column
        found_words = ()
        for comp in col_name.split():
            if comp not in seq:
                #print "not found ", comp
                continue
            else:    
                found_words = found_words + (comp,)
                
        #inspect if found all...
        if found_words == tuple(col_name.split()):
            #print "found all words ", found_words, " t ", tuple(col_name.split()), " candidate_index ", iter
            candidate_index = iter
            
    return candidate_index
```

```python
def employee_line_parser(item):
    
    resd = {'Unknown':''}
    #don't waste time on empty strings
    if item is None or item.isspace():
        return resd
    
    dd = {
    'Social Security' : lambda y: re.search(r'\d{3}-\d{2}-\d{4}',item),
    'City'            : lambda y: re.match(r'([a-zA-Z].*)[,]*[ ]([N][YJ]|[0-9]{5})',item), 
    'State'           : lambda y: re.match(r'([a-zA-Z].*)[,]*[ ]([N][YJ]|[0-9]{5})',item),
    'Employee Name'   : lambda y: re.match(r'([a-zA-Z]+)[ .-]+([a-zA-Z]+)(\D*)$',item),
    'Street Name'     : lambda y: re.match(r'(\d*)\s*[a-zA-Z0-9].*(ST|AVE|RD|LA|CT|PL|BLVD|PKWY|STREET|AVENUE|ROAD|LANE|COURT|PLACE|BOULEVARD|PARKWAY)[ ]*$',item),
    }
   
    for key in dd.keys():
        # test if tag is in the dictionary; test if value matches the criteria
        if dd[key](item):
            # Return whole or substring, depending if city/state
            if key in ('City','State'):
                #stripping pesky comma from city... fix with group
                resd = {'City':dd[key](item).group(1).strip(','),'State':dd[key](item).group(2)}
            else:
                resd = {key:dd[key](item).group(0)}
            break

    return resd
```

```python
def employee_record(sheet,row,column):
    employee = tuple(sheet.cell_value(row+i,column) for i in range(0,4))
    loc = dict(sheet=sheet.name, row=row, column=column)
 
    ei = dict((x,'') for x in ('Social Security','City','State','Employee Name','Street Name'))
    
    for item in employee:
        #f_val returns dictonary {'State':'NY', etc }
        vals = employee_line_parser(item)
        if vals is not None:
            for key,val in vals.items():
                ei[key] = val

    return {'employee raw data': employee, 'record location': loc, 'employee data': ei }
```

```python
def end_of_table(sheet,row):

    key_words = ("FALSIFICATION OF STATEMENT","I HEREBY CERTIFY","I UNDERSTAND THAT")
    
    for row in tuple(sheet.row_values(row+i) for i in range(0,4)):
        for cell in row:
            for key in key_words:
                if key in str(cell):
                    return True
    return False            
```

```python
def employee_record_parser(book,sheet):

    find_first_employee_record_in_table = True
    employees = list()
    r=0
    #each loop iteration processes 4 rows at a time, hence the while will stop on the 4th to the last row
    while r < sheet.nrows-3:
        
        if find_first_employee_record_in_table == True:
            if sheet.cell_type(r,0) == xlrd.XL_CELL_EMPTY or sheet.cell_type(r,0) == xlrd.XL_CELL_BLANK:
                r=r+1
            elif sheet.cell_type(r,0) != xlrd.XL_CELL_TEXT:
                #print "Trouble for EMPLOYEES NAME", sheet.name, r, sheet.cell_type(r,0), xlrd.XL_CELL_BLANK, sheet.cell_value(r,0)
                r=r+1
            elif ("EMPLOYEES NAME") in sheet.cell_value(r,0) or ("NAME, ADDRESS AND") in sheet.cell_value(r,0):
                #The first record is under a first line aftr EMPLOYEES NAME
                #The first line of the employee record can be empty
                cell = sheet.cell(r,0)
                fmt1 = book.xf_list[sheet.cell(r+1,0).xf_index]
                fmt2 = book.xf_list[sheet.cell(r+2,0).xf_index]
                fmt3 = book.xf_list[sheet.cell(r+3,0).xf_index]
             
                if fmt1.border.bottom_line_style != 0 or fmt2.border.top_line_style !=0: r=r+2
                elif fmt2.border.bottom_line_style != 0 or fmt3.border.top_line_style !=0: r=r+3
                
                find_first_employee_record_in_table = False
                #print "Found first employee record at ", r
            else:
                #just increment
                r=r+1
        else:
            #test if end of table
            if end_of_table(sheet,r) == True:
                find_first_employee_record_in_table = True
                #print "End of table detected at ", r, " with ", key
            else:
                #cleared for employee record
                employees.append(employee_record(sheet,r,0))
                r = r+4
                #print "New r is ", r
                        
    return (employees)
```

```python
def hrs_record_parser(sheet,row,col):
    hrs = [cell(sheet.cell_value(row+i,col),sheet.cell_type(row+i,col)) for i in range (0,4)]

    return sum(hrs)
```
